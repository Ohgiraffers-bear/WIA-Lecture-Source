# 13. 원 그래프 (Pie Chart)
---
### 원 그래프의 특징
- 원 그래프는 전체에서 각 항목이 차지하는 비율을 시각화하는 데 효과적이다.
- 각 부분의 크기를 직관적으로 비교할 수 있다.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 사용자
matplotlib.rcParams['font.size'] = 15 # 폰트 크기 설정
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
```

## 기본 원 그래프
### pie() 함수
- **pie()** 함수를 사용하여 원 그래프를 그린다.
- **autopct**: 백분율 표시 형식
- **startangle**: 시작 각도
- **counterclock**: 반시계 방향 여부


```python
brands = ['삼성', '애플', 'LG', '기타']
market_share = [45, 40, 25, 10] 

plt.pie(market_share, labels=brands, autopct='%.1f%%',startangle=90, counterclock=False) # %로 표시가됨.

plt.show()
```

## 조각 분리하기
### explode 파라미터
- **explode**를 사용하여 특정 조각을 중심에서 분리할 수 있다.
- 강조하고 싶은 항목을 돋보이게 만들 수 있다.


```python
brands = ['삼성', '애플', 'LG', '기타']
market_share = [45, 40, 25, 10] 
explode = [0.1, 0.1, 0, 0]
plt.pie(market_share, labels=brands, autopct='%.1f%%',
        startangle=90, counterclock=False, explode=explode) # %로 표시가됨.

plt.show()
```

### 모든 조각 분리
- 모든 조각을 동일하게 분리할 수도 있다.


```python
brands = ['삼성', '애플', 'LG', '기타']
market_share = [45, 40, 25, 10] 
explode = [0.1, 0.1, 0.1, 0.1]
plt.pie(market_share, labels=brands, autopct='%.1f%%',
        startangle=90, counterclock=False, explode=explode) # %로 표시가됨.

plt.show()
```

## 범례 추가하기
### legend() 함수
- 원 그래프에도 범례를 추가할 수 있다.
- **loc**: 범례 위치 지정
- **title**: 범례 제목


```python
brands = ['삼성', '애플', 'LG', '기타']
market_share = [45, 40, 25, 10] 
explode = [0.1, 0.1, 0.1, 0.1]
plt.pie(market_share, labels=brands, autopct='%.1f%%',
        startangle=90, counterclock=False, explode=explode) # %로 표시가됨.

plt.legend(loc=(1.2, 0.3), title='브랜드별 점유율')
plt.show()
```
